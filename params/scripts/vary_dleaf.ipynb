```python
import xarray as xr
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import netCDF4
import os
%matplotlib inline
```

```python
basepftfile = '/glade/p/cgd/tss/people/oleson/inputdata/lnd/clm2/paramdata/clm5_params.c171117.nc'
```

```python
defaults = xr.open_dataset(basepftfile)
```

```python
ixpft = 7
defaults['pftname'][ixpft]
defaults['dleaf'][ixpft]
```

```python
dleaf_values = np.linspace(0.0081,0.243,10)
```

```python
prefix = 'vary_dleaf_'
first = 0
i = 0
for v in dleaf_values:
    i +=1
    pftfile = "../paramfiles/spinme/"+prefix+str(first+i).zfill(4)+".nc"
    print(pftfile)
    cmd = 'cp '+basepftfile+' '+pftfile
    os.system(cmd)
    if v!='default':
        dset = netCDF4.Dataset(pftfile,'r+')
        dset['dleaf'][:]=v
        dset.close()
```
